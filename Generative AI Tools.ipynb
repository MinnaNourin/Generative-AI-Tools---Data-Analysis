# Column Describtion

* tool_name – Name of the AI tool or platform.
* company – Developer or organization behind the tool.
* category_canonical – Main purpose or type of the tool (e.g., LLMs, Image Gen).
* modality_canonical – Type of data the tool works with (text, image, multimodal).
* open_source – Indicates if the tool is open-source (1) or not (0).
* api_available – Shows if an API is available (1 = yes, 0 = no).
* api_status – Current availability of the API.
* website – Official website link of the tool.
* source_domain – Domain extracted from the website URL.
* release_year – Year the tool was launched.
* years_since_release – Number of years since the tool was released.
* mod_text – Supports text-based tasks.
* mod_image – Supports image-related tasks.
* mod_video – Supports video tasks or generation.
* mod_audio – Works with sound, voice, or music.
* mod_code – Assists with coding or programming tasks.
* mod_design – Supports design or creative tasks.
* mod_infra – Helps with AI infrastructure or deployment.
* mod_productivity – Improves workflow or productivity.
* mod_safety – Focuses on AI safety or ethical use.
* mod_multimodal – Handles multiple data types at once.
* modality_count – Total number of modalities supported by the tool.

# Visualization
**Aim:To visualize the evolution of Generative AI tools over time, highlighting growth trends, popular categories,
and the rise of open-source and multimodal technologies**

## Trend of  AI Tool Releases Over Years

```python
# Count number of tools released per year
year_count = data['release_year'].value_counts().sort_index()
year_count
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x=year_count.index, y=year_count.values, marker='o', linewidth=2, color='teal')
plt.title("Number of Tools Released per Year", fontsize=14)
plt.xlabel("Release Year", fontsize=12)
plt.ylabel("Number of Tools", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# Insights
* The graph shows how many AI tools were released each year.  
* There’s a steady <b>increase in AI tool development, especially in recent years.</b>  
* Peak years show when <b>AI innovation was most active.</b>
* The AI industry is <b>expanding fast — showing high market interest and investment.</b> 
* Years with fewer releases may indicate slower development or fewer startups.
* Focusing on niche or multimodal tools helps stand out in a crowded market.  
* The upward trend shows AI is a sustainable, growing field with future potential.

# AI Tools Released per Year by Category

```python
release_cat = data.groupby(['release_year', 'category_canonical']).size().unstack(fill_value=0)
release_cat.plot(kind='bar', stacked=True, figsize=(12,6), colormap='tab20')
plt.title('AI Tools Released per Year by Category', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Number of Tools')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# Insights
- Shows number of AI tools <b>released each year by category.</b>  
- Clear <b>rise in tool releases</b>, especially after **2020.**  
- <b>LLMs, Chat Assistants, and Image tools</b> lead recent years.  
- <b>Safety and Productivity tools/b> are **slowly increasing.**  
- Indicates fast growth and wider AI adoption.  
- Fewer releases in early years show AI was still developing.  
- The trend highlights where innovation and investment are most active in AI.

# Top 10 Companies by AI Tools and Supported Modalities

```python
# Define modality columns
modality_cols = ['mod_text', 'mod_image', 'mod_video', 'mod_audio', 'mod_code', 
                 'mod_design', 'mod_infra', 'mod_productivity', 'mod_safety', 'mod_multimodal']

top_companies = data['company'].value_counts().head(10).index
top_data = data[data['company'].isin(top_companies)]
modality_by_company = top_data.groupby('company')[modality_cols].sum()
professional_colors = ['#2E4057','#3E6C76','#5A5A5A', '#7D5BA6','#A3A3A3','#1F77B4','#4C566A','#6C5B7B','#8C9EA6','#345B63']
plt.figure(figsize=(12,8))
modality_by_company.plot(kind='barh',stacked=True, color=professional_colors, edgecolor='black')
plt.title("Top 10 Companies by AI Tools and Supported Modalities", fontsize=16)
plt.xlabel("Number of Tools", fontsize=12)
plt.ylabel("Company", fontsize=12)
plt.legend(title='Modality', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# Insights
- Shows how many types (modalities) of AI tools each top company supports.  
- Each color represents a different modality — like text, image, audio, or code.  
- Companies with wider bars cover more types of AI tools.  
- Some focus on text or image tools, while others build multi-modal systems.  
- Indicates how diversified or specialized each company is in AI development

## Distribution of Open-Source and Closed-Source AI Tools

```python
data['open_source'].value_counts()
# 0 - closed source
# 1 -open source
```

```python
# Count the number of open source and closed source tools
open_source_counts = data['open_source'].value_counts()
labels = ['Closed Source', 'Open Source']
colors = ['#ff9999','#66b3ff']  
explode = (0, 0.1)  
plt.figure(figsize=(6,6))
plt.pie(open_source_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, explode=explode, shadow=True)
plt.title('Open Source vs Closed Source AI Tools')
plt.show()
```

# Insights
* Shows how many tools are open or closed source.
* Most are closed-source, fewer are open-source.
* Many companies keep tools closed to protect their ideas and make money.
* Open-source tools help in research, learning, and faster innovation.



# Open Source vs Closed Source LLM Tools by Top Companies

# Insights
* This chart shows which top companies make open or closed chat and language tools.
* Open-source tools are <b>shared freely so anyone can use or improve them.</b>
* Closed-source tools are <b>kept private to protect ideas and earn money from users.</b>
* Closed-source tools are <b>secured,prevent copying, and maintain good quality.</b>
* Open-source helps with <b>collaboration and learning.</b>

# Supported Modalities Across Generative AI Tools

```python
# count api-availability
api_count = data['api_available'].value_counts()
api_count
modality_cols = ['mod_text', 'mod_image', 'mod_video', 'mod_audio', 'mod_code', 
                 'mod_design', 'mod_infra', 'mod_productivity', 'mod_safety', 'mod_multimodal']

# Count how many tools support each modality
modality_count = data[modality_cols].sum().sort_values(ascending=False)
modality_count

sns.set_style("whitegrid")
plt.figure(figsize=(10,6))
bars = sns.barplot(x=modality_count.values, y=modality_count.index, palette='crest', edgecolor='black')
for i, v in enumerate(modality_count.values):
    bars.text(v + 0.3, i, str(v), color='black', va='center', fontweight='bold')
plt.title("Supported Modalities Across Generative AI Tools", fontsize=16)
plt.xlabel("Number of Tools Supporting Each Modality", fontsize=12)
plt.ylabel("Modality Type", fontsize=12)
plt.tight_layout()
plt.show()
```

# Insights
* Shows number of AI tools <b>supporting each modality (text, image, audio, etc.).</b>
* <b>Text and image</b> are the most common modalities because <b>easier to train and more in demand.</b>
* Commonly used in <b>chatbots, design, and content tools.</b>
* <b>Audio, video, and safety tools</b> are <b>less developed.</b>
* Audio, video, and multimodal tools need<b> more data and computing power</b>.Fewer companies build them due to complexity and cost.
* Future trend: Multimodal AI will grow as technology advances.



# Category vs Modality

```python
cross_tab = pd.crosstab(data['category_canonical'], data['modality_canonical'])
plt.figure(figsize=(12,6))
sns.heatmap(cross_tab, annot=True, fmt='d', cmap='Blues')
plt.title('Category vs Modality')
plt.show()
```

# Insights
- This heatmap shows how different **AI categories** (like Chat Assistants, Image Tools, etc.) are linked with their **modalities** (Text, Image, Audio, etc.).  
- Darker boxes represent **more tools** in that category–modality combination.  
- For example, text-based modalities are dominant in **LLMs & Chat Assistants**, while **image** is more common in **Design** or **Art tools**.  
- It helps us understand **which types of AI are most developed** and where there’s potential for new innovations.
